```python
def nate(N):
    b = 1 + 1/N
    p = 1
    for i in range(0,N):
        p *= b
    return p

def exnate(N):
    return nate(N)+N*(nate(N+1)-nate(N-1))/2
```

```python
import math
N=10000
print(nate(N), exnate(N), math.exp(1))
```

```python
x=[]
y=[]
z=[]
dy=[]
dz=[]
dy_dz = []
dy_dz_y =[]
for i in range(2,1000):
    x += [i]
    y += [exnate(i)]
    dy += [exnate(i+1)-exnate(i)]
    z += [1/i]
    dz += [1/(i+1)-1/i]
    dy_dz += [dy[i-2]/dz[i-2]]
    dy_dz_y += [dy_dz[i-2]/y[i-2]]
```

```python
plot=plt.plot(x, y)  # Matplotlib plot.
```

```python
plt.plot(z, dy_dz_y)
plt.title('z-dy_dz')
plt.grid(True)
```

```python
def power(b,N):
    if N==1: return b
    d = divmod(N,2)
    p = power(b,d[0])
    if d[1]:
        a = b
    else:
        a = 1
    return p*p*a
```

```python
power (1.1,200)
```

```python
def e1(dx):
    return power(1+dx,1/dx)
```
