```python
k0=1.5
dk=1/300

def k(n):
    return k0-dk*n

n=[1]
for t in range(1,100):
    n+=[k(n[-1])*n[-1]]

import matplotlib.pyplot as plt
import numpy as np

t=list(range(0,100))

fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot(t, n)  # Plot some data on the axes.
```

Scaled by $n_\infty$
======

```python
ninf=(k0-1)/dk
ninf
```

```python
def ks(ns):
    return k0-(k0-1)*ns

n=[1/ninf]
for t in range(1,100):
    n+=[ks(n[-1])*n[-1]]

import matplotlib.pyplot as plt
import numpy as np

t=list(range(0,100))

fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot(t, n)  # Plot some data on the axes.
```

Time scale
==========

```python
Δt=.5
tc=list(map(lambda x:x*Δt,range(0,int(20/Δt))))
def kss(ns):
    return 1+(k0-1)*(1-ns)*Δt
nc=[1/ninf]
for l in range(1,int(20/Δt)):
    nc+=[kss(nc[-1])*nc[-1]]
    
from math import exp

def fncc(t):
    A=nc[0]/(1-nc[0])
    E=A*exp((k0-1)*t)
    return E/(1+E)

ncc=[fncc(t) for t in tc]

import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot(tc, nc)  # Plot some data on the axes.
ax.plot(tc,ncc)
```
