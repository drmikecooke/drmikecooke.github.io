```python
import numpy.polynomial.polynomial as p
import scipy.linalg as la
import numpy as np
```

```python
g =[0.25,0.75,1]
pg = [1,0.75,0.25,-1]
```

```python
characteristic = p.Polynomial(pg)
```

```python
croots=characteristic.roots()
```

```python
I = [1] # begin creation of infection list - 1 on day 0

forwardlist = [[1]*len(croots)] # list of zero powers of characteristic roots (i.e. 1)
backlist = forwardlist.copy()

for n in range(1,50):
    I = I + [0] # new entry
    for m in range(0,len(g)):
        if (n-m-1 >= 0): I[n] = I[n] + g[m]*I[n-m-1] # ensure that list indices remain non-negative
    forwardlist = forwardlist + [[forwardlist[n-1][m]*croots[m] for m in range(0,len(croots))]]
    backlist = backlist + [[backlist[n-1][m]/croots[m] for m in range(0,len(croots))]]
```

```python
def vcoeff(croots):
    v=[]
    for a in croots:
        prod = 1
        for b in croots:
            if a!=b:
                prod *=b/a-1
        v += [1/prod]
    return v
    
```

```python
P = np.matrix(forwardlist)*np.matrix(vcoeff(croots)).T # prediction list

B = np.matrix(backlist)*np.matrix(vcoeff(croots)).T # prediction list # retro list
```
